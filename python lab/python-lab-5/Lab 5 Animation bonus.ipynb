```python
# Initial values
v0 = 20.0 ## Change this !!!
angle = 15.0 ## Change this !!!
t_end = 1.0 ## Change this !!!
```

```python
# Initialize random number generator with student ID
rng = np.random.default_rng(student_id)

# Initial position
x0 = 0.0
y0 = rng.uniform(0.5, 1.5)  # Initial height in m
print(f'x0 = {x0:.2f} m')
print(f'y0 = {y0:.2f} m')

# Initial velocity and angle
theta = np.deg2rad(angle)
print(f'v0 = {v0:.2f} m/s')
print(f'angle = {angle:.2f} deg')

# End of flight
print(f't_end = {t_end:.2f} s')
g = 9.81 # Gravitational constant

# Define the initial conditions and parameters for the projectile motion
# This code simulates the trajectory of a ball thrown towards a target
yl = rng.uniform(2, 3)  # Level height in m
xl0 = rng.uniform(2, 5)  # Level x position in m
xw = rng.uniform(1, 3)  # Level x width in m

# Target position
x_target = rng.uniform(0, 5) + np.array([10, 9.5, 9.5, 10, 10])
y_target = rng.uniform(0, 1) + np.array([0.5, 0.5, 0.3, 0.3, 0.5])

# Fence position
x_edge = np.array([0, xl0, xl0, xl0+xw, xl0+xw, 20])
y_edge = np.array([0, 0,   yl,  yl, 0,   0])

# Define the time vector for the trajectory
t = np.arange(0, t_end, 0.01)

# Calculate the trajectory of the projectile
x = x0 + v0 * np.cos(theta) * t
y = y0 + v0 * np.sin(theta) * t - 0.5 * g * t**2
```

```python
# Create a figure and axis for the animation
fig, ax = plt.subplots(figsize = (7, 4))
plt.sca(ax)

# Plot the fence, start and target positions
plt.plot(x_edge, y_edge, 'b--', label = 'Fence')
plt.plot(x_target, y_target, 'r-', label = 'Target')
plt.plot(x0, y0, 'g^', label = 'Start')

# Initialize the trajectory of the ball
l, = ax.plot(x[0],y[0], 'g-', label = 'Trajectory')     # Tail of the trajectory
l2, = ax.plot(x[0], y[0], 'go', label = 'Ball')   # Position of the ball
ax.set_title(f't = {t[-1]:.2f}')  # Change the title to the elapsed time

# Annotate the rest of graph
plt.xlabel('x position (m)')
plt.ylabel('y position (m)')
plt.xlim(-1, 21)
plt.ylim(-0.2, 7)
plt.legend(loc = 'upper right')
plt.grid(True)

# Animation function
def animate(i):
    i0 = max(0, i-10)               # Get index for 10 last points from the data
    l.set_data(x[i0:i], y[i0:i])    # Show the tail
    l2.set_data([x[i]], [y[i]])     # Show the position
    ax.set_title(f't = {t[i]:.2f}') # Update the title

# Create the animation
ani = FuncAnimation(fig, animate, frames=len(t), interval = 20)
plt.close(fig)

# Show animation in Javascript HTML object
html = ani.to_jshtml()
display(HTML(html))

# When ready, click the Play-button in the graphics below.
```
