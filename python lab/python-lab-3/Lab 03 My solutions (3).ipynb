### My solution

```python
# start your code here
x = np.linspace(-10,10,1000)
y = (
    ((x >= -0.4)& (x< 2.6)) *(0.4 + x)
+
    ((x >= 2.6)& (x< 5.6)) * (5.6 - x)
)

plt.figure(figsize=(10,6))
plt.plot(x, y , label='f(x)', color='orange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Problem 1: Piecewise Function')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
