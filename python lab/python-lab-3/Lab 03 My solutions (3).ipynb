### My solution

```python
# start your code here
x = np.linspace(-10,10,2000)
y = (
    ((x >= 0.8)& (x< 2.8)) *(-0.8 + x)
+
    ((x >= 2.8)& (x< 4.8)) * (4.8 - x)
)

plt.figure()
plt.plot(x, y , label='f(x)', color='orange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Problem 1: Piecewise Function')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
