```python
# Your code starts here
plt.figure(figsize=(10,7))
x = np.linspace(0, 10,400)
f = np.sqrt(2.4 * x)
g = (x-3.42)**2
plt.plot(x,f, label="$f(x) = sqrt(2.4x)$")
plt.plot(x,g, label="$g(x) = (x-3.42)**2$")
plt.title("Overview: Both functions")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.xlim(1.90, 1.97)
plt.ylim(2, 3)
plt.title("Intersection Point 1 (zoomed)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.xlim(5.22, 5.5)
plt.ylim(3,4)
plt.title("Intersection Point 2 (zoomed)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
