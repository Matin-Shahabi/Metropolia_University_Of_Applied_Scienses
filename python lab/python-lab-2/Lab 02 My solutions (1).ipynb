```python
# Your code starts here
x = np.linspace(0, 10,500)
k = 4.5
c = 5.16
f = np.sqrt(k * x)
g = (x-c)**2
plt.figure()

plt.plot(x,f, label="$f(x) = sqrt(4.5x)$")
plt.plot(x,g, label="$g(x) = (x-5.16)**2$")
plt.title("Overview: Both functions")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.xlim(2, 4)
plt.ylim(1, 5)
plt.title("Intersection Point 1 (zoomed)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

plt.figure()
plt.plot(x, f, label='f(x)')
plt.plot(x, g, label='g(x)')
plt.xlim(6, 8)
plt.ylim(4,7)
plt.title("Intersection Point 2 (zoomed)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
