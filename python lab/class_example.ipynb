```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v0_kmh = 75
v0 = v0_kmh*1000/3600
t = np.arange(0,30,0.1)
x = v0 * t
plt.plot(t,x)
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.grid()
plt.show()
```

```python
po_v0 = 2.5
x2 = 1/2*po_v0*t**2
plt.plot(t,x2)
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.title(f"Police car accelerating at {po_v0} m/s")
plt.grid()
plt.show()
```

```python
plt.plot(t,x2,label=f'Police car , {po_v0} m/s')
plt.plot(t,x,label=f'Motorist , {v0} m/s')
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.legend()
plt.grid()
plt.show()
```

```python
i = (x2 > x) #boolean mark for the travelled distance of tye police > motorist
plt.plot(t,i)
plt.xlabel('Time (s)')
plt.ylabel('x2 > x')
plt.grid()



time_catch = np.min(t[i])
plt.title(f"Time of catch : {time_catch} s")
plt.show()
```

```python
i2= (t==time_catch)
x0 = x[i2]
plt.plot(t,x2,label=f'Police car , {po_v0} m/s')
plt.plot(t,x,label=f'Motorist , {v0} m/s')
plt.axvline(time_catch , color='red',label=f't = {time_catch} s')
plt.axhline(x0, color='red',label=f'x = {x0[0]:.0f} m')
plt.xlabel('Time (s)')
plt.ylabel('Distance (m)')
plt.legend()
plt.grid()
plt.show()
```

```python
v_po = po_v0*t*3600/1000
i2 = (t==time_catch)
v2 = v_po[i2]
plt.plot(t,v_po)
plt.axvline(time_catch,color = "red")
plt.axhline(v2,color = "red")
plt.xlabel('Time (s)')
plt.xlabel('Time (hours)')
plt.ylabel('Velocity (km)')
plt.title(f'The speed of the police car when catch motorist is {v2} km/h')
plt.grid()
plt.show()
```

```python
x0,y0 = 0,1.5 #starting position of the ball
v0 = 25 #m/s
theta = np.deg2rad(30) #convert from degree to radian
g = 9.81
t = np.arange(0,3,0.001) #estimate when the ball reaches the floor
vx = v0*np.cos(theta)
vy = v0*np.sin(theta) - g*t
x = x0 + vx*t # position of the ball horizontally
y = y0 + vy*t - 1/2*g*t**2 # position of the ball vertically
plt.plot(x,y)
plt.axhline(0,color='red')
plt.axvline(0,color='red')
plt.grid()
plt.show()


```

```python
i = (y<= 0)
plt.plot(t,i)
plt.grid()
plt.show()
t_end = np.min(t[i])
print(f't_end: {t_end} s')
```

```python
i_end = (t== t_end)
x_end = x[i_end]
print(f'x_end: {x_end} m')
```
